```python
import requests
import os
import json
from lxml import etree
```

```python
files = {'file': open('data/test_plain.zip', 'rb')}
data = {'nlp_pipeline': '[["spacy", {"language": "de"}]]', "zip_type": "zip", "file_type": "text/plain", "out_format": "application/json+acdhlang"}
```

```python
files = {'file': open('data/editions.zip', 'rb')}
```

```python
url = "http://127.0.0.1:8000/query/nlppipeline-api/"
```

```python
res = requests.post(url, files=files, data=data)
```

```python
res.json()
```

```python
res2 = requests.get("http://127.0.0.1:8000/query/downloadlink-api/", params={'proc_id': res.json()['proc_id']})
```

```python
res2.json()
```

```python
dirname = res2.json()['path'].split('/')[1].split('.')[0]
lst = []
for f in os.listdir(os.path.join('tmp', dirname)):
    print(dirname)
    with open(os.path.join('tmp', dirname, f), 'r') as f2:
        r = json.load(f2)
        lst.append(r)
```

```python
lst
```
